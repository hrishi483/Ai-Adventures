```python
# FROG LEAP PUZZLE
def move_right(area,index,step):
    if step<=2 and area[index+step]=='_':
        area[index],area[index+step]=area[index+step],area[index]
    else:
        print("Invalid") 
    return area

def move_left(area,index,step):
    if step<=2 and area[index-step]=='_':
        area[index],area[index-step]=area[index-step],area[index]
    else:
        print("Invalid")
    return area

def check_right(area, index, step):
    if index + step <= len(area)-1:
        if area[index + step] == '_':
            return True
        else:
            return False
    return False

def check_left(area, index, step):
    if index - step >= 0:
        if area[index - step] == '_':
            return True
        else:
            return False
    return False


def check_fail(area):
    stop=0
    for i in range(len(area)):
        direction=area[i]
        global boo

        if direction == "R":
            for step in range(1, 3):
                boo = check_right(area, i, step)
                # stop+=1 if boo==True else stop+=0
                if boo==False:
                    stop+=0
                else:
                    stop+=1
        elif direction == "L":
            for step in range(1, 3):
                boo = check_left(area, i, step)
                # stop+=1 if boo==True else stop+=0
                if boo==False:
                    stop+=0
                else:
                    stop+=1

    if stop==0:
        print("RESET")
    else:
        print("Continue")

        
def check_win(area):
    for i in range(len(area)):
        if area[i]=='_':
            index=i
            break
    if index==3 and area[0]=="L" and area[1]=="L" and area[2]=="L":
        return True

    
def print_emoji(area):
    for i in area:
        if i=="R":
            print("\U0001F438",end=" ")
        elif i=="L" :
            print("\U0001F437",end=" ")
        elif i=='_':
            print("\U0001FAA8",end=" ")
    print("\n")
    
stop=True
area=["R","R","R","_","L","L","L"]

print(['1','2','3','4','5','6','7'])
print_emoji(area)
       
while stop:
    direction=input("Enter Direction ")
    num=int(input("Enter frog index "))
    
    if direction=="r":
        step=int(input("Enter number of steps "))
        area=move_right(area,num-1,step)
        
    elif direction=="l":
        step=int(input("Enter number of steps "))
        area=move_left(area,num-1,step)
    print_emoji(area)
    print([1,2,3,4,5,6,7])
    print("*"*50)
    
    if check_win(area):
        print("WIN")
        break
    elif check_fail(area)==True:
        print("RESET")
        break
    
```
